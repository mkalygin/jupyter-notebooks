```python
import numpy as np
import math
import scipy.stats as sts
import matplotlib.pyplot as plt
%matplotlib inline
```

#Generates the exponential distribution. 

```python
n = [5., 10., 50., 100., 250.]
colors = ['red', 'yellow', 'green', 'blue', 'purple']
l = 0.5
mu = 1./l
variance = l**(-2)


exp_rv=sts.expon(0, scale = 1./l)
arg = np.linspace(0, 8, 50)
val = exp_rv.pdf(arg)
```

###$\mu$ = $1/\lambda$
##$\sigma$ = 1/$\lambda$*1/$\sqrt(n)$

##Plots the exponential distribution and the hystogram with chosen values.

```python
plt.plot(arg, val)
plt.ylabel('$y$')
plt.xlabel('x')
sample = exp_rv.rvs(1000)
hist = plt.hist(sample, bins = 40, normed = True)
```

##Hystogram and theoretical normal distribution of average value for different sampe volumes:

###N = 5

```python
arg = np.linspace(-1, 8, 80)
norm_rv = sts.norm(loc = mu, scale = (variance/n[0])**(1./2))
val = norm_rv.pdf(arg)
plt.plot(arg, val)
plt.ylabel('Average')
plt.xlabel('x')
avg = []
for i in range(1000):
    avg.append(np.mean(exp_rv.rvs(n[0])))
graph1 = plt.hist(avg, bins = 50, color = colors[0], normed = True)
```
